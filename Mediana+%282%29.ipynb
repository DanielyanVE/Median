```python
def fn(a, b):
    if np.mean(a==b) != 1. :
        idx = np.where( (a>b) != (a<b) )[0][0]
        if a[idx] < b[idx]: return True
        if a[idx] > b[idx]: return False
```

```python
def float_to_binary(num):
    exponent=0
    shifted_num=num
    while shifted_num != int(shifted_num):        
        shifted_num*=2
        exponent+=1
    if exponent==0:
        return '{0:0b}'.format(int(shifted_num))
    binary='{0:0{1}b}'.format(int(shifted_num),exponent+1)
    integer_part=binary[:-exponent]
    fractional_part=binary[-exponent:].rstrip('0')
    k = ''
    for i in fractional_part:
        k+=i+" "
    return np.fromstring(k,dtype=int,sep=' ')[:14]
```

```python
def somefunc(x):
    b = np.random.randint(2,size = (x,14))
    z = np.apply_along_axis(fn,1,b,c)
    return np.mean(z)
somefunc = np.vectorize(somefunc)
```
